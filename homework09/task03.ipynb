```python
#3. Посчитать коэффициент линейной регрессии при заработной плате (zp),используя градиентный спуск (без intercept).
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
zp = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
```

```python
ks = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])
```

```python
step = 1e-6
```

```python
n = len(zp)
b1 = 0.1
b0 = 500
```

```python
def mse_(b0, b1, y=ks, x=zp, n=n):
    return np.sum((b0 + b1 * x - y) ** 2) / n
```

```python
for i in range(10000):
    b1 -= step * (2/n)*np.sum((b1*zp-ks)*zp)
    b0 -= step * (2/n)*np.sum((b0*zp-ks))
    if i % 100 == 0:
        print(f'iteration: {i}, b0: {b0}, b1 : {b1}, mse: {mse_(b0, b1)}')
```
